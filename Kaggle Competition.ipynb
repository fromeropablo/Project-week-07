## Reassign values to the column cut

```python
dic_para_hot = { "Premium": 0.9,
                "Very Good": 0.7,
                "Good": 0.5,
                "Ideal":0.3,
                "Fair":0.1,
                
}
data.cut = data.cut.map(dic_para_hot)
```

```python
data.sample(10)
```

## Reassign values to the column color

```python
dic_para_hot = { "D": 1,
                "E": 0.85,
                "F": 0.7,
                "G":0.55,
                "H":0.4,
                "I":0.25,
                "J":0.1
}
data.color = data.color.map(dic_para_hot)
```

```python
data.sample(10)
```

## Reassign values to the column clarity

```python
dic_para_hot = {"F":1 
                "IF": 1,
                "VVS1": 0.87,
                "VVS2": 0.74,
                "VS1":0.61,
                "VS2":0.48,
                "SI1":0.35,
                "SI2":0.22,
                "I1":0.09
}
data.clarity = data.clarity.map(dic_para_hot)
```

```python
data.drop(columns = ["x", "y", "z"], inplace = True)
```

```python
data
```

```python
X = data.drop(columns = ["id", "price"])
y = data.price
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Asigno las variables train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state = 666)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python

# Entrenamos el modelo
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test, y_pred))
print("MSE: ", metrics.mean_squared_error(y_test,y_pred))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R2: ", metrics.r2_score(y_test,y_pred))
```
